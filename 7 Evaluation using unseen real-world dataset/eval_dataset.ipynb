```python
import pandas as pd
import numpy as np
import pickle
from copy import deepcopy
import spacy 
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import joblib
from os import walk

class evaluate_model:
    
    
    def __init__(self, eval_tasks, model):
        
        self.target_tasks = eval_tasks['task_complete']
        self.model = model
        
    def word2vec(self):
        
        tasks_train_w2v = [row for row in self.target_tasks]
        vector_size = 100
        w2v_model_tasks = Word2Vec(tasks_train_w2v, sg = 0, size = vector_size, window=6, min_count=1, workers=4, iter= 1000)
        
        return w2v_model_tasks, vector_size
    
    @staticmethod
    def make_feature_vec(words, w2v_model, emb_size):
        """
        Average the word vectors for a set of words
        """
        feature_vec = np.zeros((emb_size,),dtype="float32")  # pre-initialize (for speed)
        nwords = 0.
        index2word_set = set(w2v_model.wv.index2word)  # words known to the model

        for word in words:
            if word in index2word_set: 
                nwords = nwords + 1.
                feature_vec = np.add(feature_vec,w2v_model[word])

        nwords = np.array([nwords])

        feature_vec = feature_vec / nwords[:,None]
        return feature_vec  
    
    def get_avg_feature_vecs(self,tasks, w2vmodel, emb_size):
        """
        Calculate average feature vectors for all tasks
        """
        counter = 0
        tasks_feature_vecs = np.zeros((len(tasks),emb_size), dtype='float64')  # pre-initialize (for speed)

        for task in tasks:
            tasks_feature_vecs[counter] = self.make_feature_vec(task, w2vmodel, emb_size)
            counter = counter + 1
            
        return tasks_feature_vecs
    
    def prepare_val_tasks(self, w2v_model_tasks, emb_size):        
        
        clean_test_tasks = [task for task in self.target_tasks]
        
        print('The total amount of tasks is ', len(clean_test_tasks))

        valDataVecs = self.get_avg_feature_vecs(clean_test_tasks, w2v_model_tasks, emb_size)
        
        return valDataVecs
    
    def run_model(self):
        
        word2vec, emb_size = self.word2vec()
        val_data_vecs = self.prepare_val_tasks(word2vec, emb_size)
        print(type(val_data_vecs))
        print('------------------')
        result = self.model.predict(val_data_vecs)
        
        return result      
        
```

```python
def get_filtered_validate(mt,mod):
    
    mt = mt[mt['mod_keys_found_string'] == mod]
    
    return mt

def get_model_from_dir(file_list, mod):
    
    ret_file = ''
    for file in file_list:
        if mod in file:
            ret_file = file
    
    return ret_file   
    

if __name__ == "__main__":
    
    with open('val_set.pkl', 'rb') as input_file:
        val_set = pickle.load(input_file)
        
    val_set['mod_keys_found_string'] = val_set['mod_keys_found'].apply(lambda x: ''.join(x))

    with open('top10_list.pkl', 'rb') as input_file:
        top10_list = pickle.load(input_file)
        
    model_files = []
    for (dirpath, dirnames, filenames) in walk('rf_models'):
        model_files.extend(filenames)
    
    result_df = pd.DataFrame()
    
    for module in top10_list:
        
        print('Evaluating RF for the module ', module)
        
        filtered_eval = get_filtered_validate(val_set, module)
        print('The size of the module ', module,' is ', filtered_eval.shape)
        
        model_file = get_model_from_dir(model_files,module)
        print('The model file is ', model_file)
        
        rf_model = joblib.load('rf_models/'+ model_file)
    
        mod_eval_obj = evaluate_model(filtered_eval,rf_model)
        
        res = mod_eval_obj.run_model()
        filtered_eval['pred_labels'] = res
        
        result_df = pd.concat([result_df,filtered_eval], axis =0)
        
        
```
